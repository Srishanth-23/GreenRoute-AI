```python
# ======================================================
# ROUTE AGENT MVP — ECO ROUTE SELECTION + GIS VISUALIZATION
# ======================================================

import pandas as pd
import numpy as np
from math import radians, sin, cos, sqrt, atan2
import folium
import random
import webbrowser
import os

# ---------- FILE PATH ----------
BASE = r"C:\Users\ASUS\My_Learning\sat hack"  # Change if folder differs

# ---------- LOAD CSV DATA ----------
trips = pd.read_csv(f"{BASE}\\trips_india.csv", parse_dates=['start_time','end_time','created_at'])
vehicles = pd.read_csv(f"{BASE}\\vehicle_agent_india_full.csv")
environment = pd.read_csv(f"{BASE}\\environment_agent_india_full.csv", parse_dates=['time_window_start','time_window_end','last_updated'])
routes = pd.read_csv(f"{BASE}\\main_agent_route_manager_india.csv", parse_dates=['last_updated'])
fuel_preds = pd.read_csv(f"{BASE}\\fuel_predictions_india.csv", parse_dates=['created_at'])

# ---------- STRIP WHITESPACE IN IDS ----------
for df in [trips, routes, fuel_preds, environment, vehicles]:
    for col in ['route_id','vehicle_id']:
        if col in df.columns:
            df[col] = df[col].astype(str).str.strip()

# ---------- HELPER FUNCTIONS ----------
def haversine_km(lat1, lon1, lat2, lon2):
    R = 6371.0
    lat1r, lon1r, lat2r, lon2r = map(radians, (lat1, lon1, lat2, lon2))
    dlat, dlon = lat2r - lat1r, lon2r - lon1r
    a = sin(dlat/2)**2 + cos(lat1r)*cos(lat2r)*sin(dlon/2)**2
    return R * 2 * atan2(sqrt(a), sqrt(1-a))

def get_vehicle_info(vehicle_id):
    row = vehicles[vehicles['vehicle_id'] == vehicle_id]
    if len(row) == 0:
        return None
    return row.iloc[0].to_dict()

def get_env_for_route(route_id):
    matches = environment[environment['route_id'] == route_id]
    if matches.empty:
        return None
    return matches.sort_values('last_updated', ascending=False).iloc[0].to_dict()

def get_fuel_prediction(route_id, vehicle_id=None):
    df = fuel_preds[fuel_preds['route_id'] == route_id]
    if vehicle_id:
        df = df[df['vehicle_id'] == vehicle_id]
    if df.empty:
        return None
    return df.sort_values('created_at', ascending=False).iloc[0].to_dict()

def estimate_fuel_fallback(distance_km, baseline_kmpl=10, multiplier=1.0, env_factor=1.0):
    return distance_km / baseline_kmpl * multiplier * env_factor

def compute_ecoscore(predicted_fuel_l, predicted_co2_kg, est_time_min, ref_time_min):
    fuel_score = 1.0 / (1.0 + predicted_fuel_l)
    co2_score = 1.0 / (1.0 + predicted_co2_kg)
    time_ratio = est_time_min / (ref_time_min + 1e-6)
    time_score = 1.0 / (1.0 + max(0, time_ratio - 1))
    return float((0.5*fuel_score + 0.3*co2_score + 0.2*time_score) * 100)

# ---------- GENERATE COORDINATES IF MISSING ----------
for col in ['start_lat','start_lon','end_lat','end_lon']:
    if col not in routes.columns:
        routes[col] = np.nan

for idx, row in routes.iterrows():
    if pd.isna(row['start_lat']) or pd.isna(row['start_lon']):
        routes.at[idx,'start_lat'] = random.uniform(8.0, 28.0)
        routes.at[idx,'start_lon'] = random.uniform(68.0, 97.0)
    if pd.isna(row['end_lat']) or pd.isna(row['end_lon']):
        routes.at[idx,'end_lat'] = random.uniform(8.0, 28.0)
        routes.at[idx,'end_lon'] = random.uniform(68.0, 97.0)

route_coords = routes.set_index('route_id')[['start_lat','start_lon','end_lat','end_lon']]
trips[['start_lat','start_lon','end_lat','end_lon']] = trips['route_id'].map(route_coords.to_dict(orient='index')).apply(pd.Series)

# ---------- ROUTE SELECTION ----------
def choose_best_route_for_trip(trip_row):
    trip_id = trip_row['trip_id']
    vehicle_id = trip_row.get('vehicle_id')

    start_lat = trip_row['start_lat']
    start_lon = trip_row['start_lon']
    end_lat = trip_row['end_lat']
    end_lon = trip_row['end_lon']

    if pd.isna(start_lat) or pd.isna(start_lon) or pd.isna(end_lat) or pd.isna(end_lon):
        raise ValueError(f"Trip {trip_id} missing coordinates!")

    routes_copy = routes.copy()
    routes_copy['dist_start'] = routes_copy.apply(
        lambda x: haversine_km(start_lat, start_lon, x['start_lat'], x['start_lon']), axis=1
    )
    candidates = routes_copy.nsmallest(10, 'dist_start')

    vehicle_info = get_vehicle_info(vehicle_id)
    ref_time = candidates['duration_min'].min() if 'duration_min' in candidates.columns else 30

    results = []
    for _, r in candidates.iterrows():
        rid = r['route_id']
        fuel_pred = get_fuel_prediction(rid, vehicle_id)
        env_info = get_env_for_route(rid)

        if fuel_pred:
            predicted_fuel = fuel_pred['predicted_fuel_l']
            predicted_co2 = fuel_pred['predicted_co2_kg']
        else:
            distance_km = r.get('route_distance_km', 10)
            baseline_kmpl = vehicle_info.get('baseline_kmpl', 10) if vehicle_info else 10
            env_factor = env_info.get('fuel_modifier',1.0) if env_info else 1.0
            predicted_fuel = estimate_fuel_fallback(distance_km, baseline_kmpl, 1.0, env_factor)
            predicted_co2 = predicted_fuel * 2.31

        est_time = r.get('duration_min', ref_time)
        ecoscore = compute_ecoscore(predicted_fuel, predicted_co2, est_time, ref_time)

        results.append({
            'route_id': rid,
            'predicted_fuel_l': predicted_fuel,
            'predicted_co2_kg': predicted_co2,
            'est_time_min': est_time,
            'ecoscore': ecoscore,
            'start_lat': r['start_lat'], 'start_lon': r['start_lon'],
            'end_lat': r['end_lat'], 'end_lon': r['end_lon']
        })

    res_df = pd.DataFrame(results).sort_values('ecoscore', ascending=False).reset_index(drop=True)
    best = res_df.iloc[0].to_dict()
    return best, res_df

# ---------- VISUALIZATION (SAVE HTML & OPEN IN BROWSER) ----------
def visualize_route_web(trip_row, best_route):
    start = [best_route['start_lat'], best_route['start_lon']]
    end = [best_route['end_lat'], best_route['end_lon']]
    mid = [(start[0]+end[0])/2, (start[1]+end[1])/2]

    m = folium.Map(location=mid, zoom_start=6)
    folium.Marker(start, popup="Start", icon=folium.Icon(color="green")).add_to(m)
    folium.Marker(end, popup="End", icon=folium.Icon(color="red")).add_to(m)
    folium.PolyLine([start, end], color="blue", weight=4, opacity=0.7).add_to(m)

    map_file = os.path.join(BASE, f"trip_{trip_row['trip_id']}_map.html")
    m.save(map_file)
    webbrowser.open(f"file:///{map_file.replace(os.sep, '/')}")
    print(f"✅ Map saved and opened in browser: {map_file}")

# ---------- RUN DEMO ----------
trip_row = trips.iloc[0].to_dict()
best_route, all_routes = choose_best_route_for_trip(trip_row)
visualize_route_web(trip_row, best_route)
```

```python
import osmnx as ox
import networkx as nx
import pandas as pd
import folium
import random
import webbrowser
import os

# ---------- FILE PATH ----------
BASE = r"C:\Users\ASUS\My_Learning\sat hack"

# ---------- LOAD CSV DATA ----------
trips = pd.read_csv(f"{BASE}\\trips_india.csv")
vehicles = pd.read_csv(f"{BASE}\\vehicle_agent_india_full.csv")

# ---------- EXAMPLE CITY COORDINATES ----------
city_coords = {
    "Delhi": (28.6139, 77.2090),
    "Coimbatore": (11.0168, 76.9558),
    "Mumbai": (19.0760, 72.8777),
    "Pune": (18.5204, 73.8567),
    "Chennai": (13.0827, 80.2707)
}

# ---------- PICK ONE TRIP ----------
trip_row = trips.iloc[0]
vehicle = vehicles[vehicles['vehicle_id'] == trip_row['vehicle_id']].iloc[0]

origin_city = vehicle["city"]
destination_city = "Coimbatore"  # example
origin_point = city_coords[origin_city]
destination_point = city_coords[destination_city]

# ---------- BUILD ROAD NETWORK ----------
print("Building road network from OSM... this may take a while.")
G = ox.graph_from_place([origin_city, destination_city], network_type="drive")

# Nearest nodes to start/end points
orig_node = ox.distance.nearest_nodes(G, origin_point[1], origin_point[0])
dest_node = ox.distance.nearest_nodes(G, destination_point[1], destination_point[0])

# ---------- OPTIONAL: ECO WEIGHT ----------
fuel_efficiency = vehicle.get('actual_kmpl', 10)
emission_factor = vehicle.get('emission_factor', 2.31)
for u, v, k, data in G.edges(keys=True, data=True):
    distance_km = data.get("length", 1)/1000
    fuel_cost = distance_km / max(fuel_efficiency, 0.1)
    co2_cost = fuel_cost * emission_factor
    data["eco_weight"] = fuel_cost + co2_cost

# ---------- COMPUTE ROUTE USING NETWORKX ----------
route_nodes = nx.shortest_path(G, orig_node, dest_node, weight="eco_weight")

# ---------- CONVERT ROUTE TO LAT/LON ----------
route_coords = [(G.nodes[n]['y'], G.nodes[n]['x']) for n in route_nodes]

# ---------- VISUALIZE WITH FOLIUM ----------
midpoint = [(origin_point[0]+destination_point[0])/2,
            (origin_point[1]+destination_point[1])/2]
m = folium.Map(location=midpoint, zoom_start=6)

# Route line
folium.PolyLine(route_coords, color="blue", weight=4, opacity=0.7, tooltip="Eco Route").add_to(m)

# Start/End markers
folium.Marker(origin_point, tooltip="Origin", icon=folium.Icon(color="green")).add_to(m)
folium.Marker(destination_point, tooltip="Destination", icon=folium.Icon(color="red")).add_to(m)

# Save & open
map_file = os.path.join(BASE, f"trip_{trip_row['trip_id']}_map.html")
m.save(map_file)
webbrowser.open(f"file:///{map_file.replace(os.sep,'/')}")
print(f"✅ Map saved and opened: {map_file}")
```
