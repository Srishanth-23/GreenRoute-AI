```python
import pandas as pd
import numpy as np

# ---------- FILE PATH ----------
BASE = r"C:\Users\ASUS\My_Learning\sat hack"

# ---------- LOAD CSV DATA ----------
trips = pd.read_csv(f"{BASE}\\trips_india.csv", parse_dates=['start_time','end_time','created_at'])
vehicles = pd.read_csv(f"{BASE}\\vehicle_agent_india_full.csv")
environment = pd.read_csv(f"{BASE}\\environment_agent_india_full.csv", parse_dates=['time_window_start','time_window_end','last_updated'])
routes = pd.read_csv(f"{BASE}\\main_agent_route_manager_india.csv", parse_dates=['last_updated'])
fuel_preds = pd.read_csv(f"{BASE}\\fuel_predictions_india.csv", parse_dates=['created_at'])
co2_aggregates = pd.read_csv(f"{BASE}\\co2_aggregates_india.csv")  # optional for validation

# ---------- DATA CLEANING / HANDLING MISSING VALUES ----------
# Vehicles
vehicles['baseline_kmpl'] = vehicles['baseline_kmpl'].fillna(10)  # default km per liter
vehicles['fuel_efficiency_l_per_km'] = 1 / vehicles['baseline_kmpl']  # convert km/l to l/km
vehicles['fuel_type'] = vehicles['fuel_type'].fillna('Petrol')

# Trips
trips['distance_km'] = trips['distance_km'].fillna(0)
trips['vehicle_id'] = trips['vehicle_id'].fillna('Unknown')
trips['driver_id'] = trips['driver_id'].fillna('Unknown')
trips['route_id'] = trips['route_id'].fillna('Unknown')

# ---------- MERGE VEHICLE DATA WITH TRIPS ----------
trips = trips.merge(
    vehicles[['vehicle_id', 'fuel_type', 'fuel_efficiency_l_per_km']],
    on='vehicle_id',
    how='left'
)

# Fill any missing vehicle info after merge
trips['fuel_type'] = trips['fuel_type'].fillna('Petrol')
trips['fuel_efficiency_l_per_km'] = trips['fuel_efficiency_l_per_km'].fillna(0.1)

# ---------- MERGE FUEL PREDICTIONS ----------
# Merge on assignment_id, driver_id, vehicle_id, route_id (trip_id not available)
trips = trips.merge(
    fuel_preds[['assignment_id', 'driver_id', 'vehicle_id', 'route_id', 'predicted_fuel_l']],
    on=['assignment_id', 'driver_id', 'vehicle_id', 'route_id'],
    how='left'
)

# ---------- CO2 FACTORS ----------
co2_factors = {
    'Petrol': 2.31,  # kg CO2 per liter
    'Diesel': 2.68,
    'CNG': 2.74,
}

# ---------- CALCULATE CO2 PER TRIP ----------
def calculate_co2(row):
    # Use predicted fuel if available
    if pd.notna(row['predicted_fuel_l']):
        fuel_used = row['predicted_fuel_l']
    else:
        fuel_used = row['distance_km'] * row['fuel_efficiency_l_per_km']
    
    factor = co2_factors.get(row['fuel_type'], 2.5)  # default if unknown
    return fuel_used * factor

trips['co2_kg'] = trips.apply(calculate_co2, axis=1)

# ---------- HANDLE ANOMALIES ----------
trips.loc[trips['anomaly_flag'] == True, 'co2_kg'] = 0

# ---------- AGGREGATE CO2 ----------
total_co2 = trips['co2_kg'].sum()
co2_by_vehicle = trips.groupby('vehicle_id')['co2_kg'].sum().reset_index().sort_values(by='co2_kg', ascending=False)
co2_by_driver = trips.groupby('driver_id')['co2_kg'].sum().reset_index().sort_values(by='co2_kg', ascending=False)
co2_by_route = trips.groupby('route_id')['co2_kg'].sum().reset_index().sort_values(by='co2_kg', ascending=False)

# ---------- SAVE RESULTS ----------
trips.to_csv(f"{BASE}\\trips_with_co2.csv", index=False)
co2_by_vehicle.to_csv(f"{BASE}\\co2_by_vehicle.csv", index=False)
co2_by_driver.to_csv(f"{BASE}\\co2_by_driver.csv", index=False)
co2_by_route.to_csv(f"{BASE}\\co2_by_route.csv", index=False)

# ---------- PRINT SUMMARY ----------
print(f"Total CO2 emitted (kg): {total_co2:.2f}")
print("\nTop 5 vehicles by CO2 emission:")
print(co2_by_vehicle.head())
print("\nTop 5 drivers by CO2 emission:")
print(co2_by_driver.head())
print("\nTop 5 routes by CO2 emission:")
print(co2_by_route.head())
```

```python
import pandas as pd
import numpy as np
from math import radians, cos, sin, asin, sqrt

# ---------- FILE PATH ----------
BASE = r"C:\Users\ASUS\My_Learning\sat hack"

# ---------- LOAD CSV DATA ----------
trips = pd.read_csv(f"{BASE}\\trips_india.csv", parse_dates=['start_time','end_time','created_at'])
vehicles = pd.read_csv(f"{BASE}\\vehicle_agent_india_full.csv")
environment = pd.read_csv(f"{BASE}\\environment_agent_india_full.csv", parse_dates=['time_window_start','time_window_end','last_updated'])
routes = pd.read_csv(f"{BASE}\\main_agent_route_manager_india.csv", parse_dates=['last_updated'])
fuel_preds = pd.read_csv(f"{BASE}\\fuel_predictions_india.csv", parse_dates=['created_at'])
co2_aggregates = pd.read_csv(f"{BASE}\\co2_aggregates_india.csv")  # optional for validation

# ---------- DATA CLEANING ----------
vehicles['baseline_kmpl'] = vehicles['baseline_kmpl'].fillna(10)
vehicles['fuel_efficiency_l_per_km'] = 1 / vehicles['baseline_kmpl']
vehicles['fuel_type'] = vehicles['fuel_type'].fillna('Petrol')

trips['distance_km'] = trips['distance_km'].fillna(0)
trips['vehicle_id'] = trips['vehicle_id'].fillna('Unknown')
trips['driver_id'] = trips['driver_id'].fillna('Unknown')
trips['route_id'] = trips['route_id'].fillna('Unknown')

environment['fuel_modifier'] = environment['fuel_modifier'].fillna(1.0)
# If distance_km does not exist, calculate it from coordinates
if 'distance_km' not in environment.columns:
    def haversine(lat1, lon1, lat2, lon2):
        lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
        dlon = lon2 - lon1
        dlat = lat2 - lat1
        a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
        c = 2 * asin(sqrt(a))
        km = 6371 * c
        return km
    environment['distance_km'] = environment.apply(
        lambda row: haversine(
            row['segment_start_lat'], row['segment_start_lon'],
            row['segment_end_lat'], row['segment_end_lon']
        ), axis=1
    )

# ---------- MERGE VEHICLE DATA ----------
trips = trips.merge(
    vehicles[['vehicle_id', 'fuel_type', 'fuel_efficiency_l_per_km']],
    on='vehicle_id', how='left'
)
trips['fuel_type'] = trips['fuel_type'].fillna('Petrol')
trips['fuel_efficiency_l_per_km'] = trips['fuel_efficiency_l_per_km'].fillna(0.1)

# ---------- MERGE FUEL PREDICTIONS (OPTIONAL) ----------
trips = trips.merge(
    fuel_preds[['assignment_id','driver_id','vehicle_id','route_id','predicted_fuel_l']],
    on=['assignment_id','driver_id','vehicle_id','route_id'],
    how='left'
)

# ---------- CO2 FACTORS ----------
co2_factors = {'Petrol': 2.31, 'Diesel': 2.68, 'CNG': 2.74}

# ---------- MERGE TRIPS WITH ENVIRONMENT SEGMENTS ----------
segments = environment.merge(
    trips[['trip_id','driver_id','vehicle_id','route_id','fuel_type','fuel_efficiency_l_per_km','predicted_fuel_l','anomaly_flag']],
    on='route_id', how='inner'
)

# ---------- CALCULATE CO2 PER SEGMENT ----------
def segment_co2(row):
    if pd.notna(row['predicted_fuel_l']):
        fuel_used = row['predicted_fuel_l']
    else:
        fuel_used = row['distance_km'] * row['fuel_efficiency_l_per_km'] * row['fuel_modifier']
    factor = co2_factors.get(row['fuel_type'], 2.5)
    return fuel_used * factor

segments['co2_kg'] = segments.apply(segment_co2, axis=1)

# Set CO2 to 0 for anomalies
segments.loc[segments['anomaly_flag']==True, 'co2_kg'] = 0

# ---------- AGGREGATE CO2 PER TRIP ----------
co2_per_trip = segments.groupby('trip_id')['co2_kg'].sum().reset_index()
trips = trips.merge(co2_per_trip, on='trip_id', how='left', suffixes=('','_so_far'))

# ---------- AGGREGATE CO2 BY ENTITY ----------
total_co2 = trips['co2_kg'].sum()
co2_by_vehicle = trips.groupby('vehicle_id')['co2_kg'].sum().reset_index().sort_values(by='co2_kg',ascending=False)
co2_by_driver = trips.groupby('driver_id')['co2_kg'].sum().reset_index().sort_values(by='co2_kg',ascending=False)
co2_by_route = trips.groupby('route_id')['co2_kg'].sum().reset_index().sort_values(by='co2_kg',ascending=False)

# ---------- SAVE RESULTS ----------
trips.to_csv(f"{BASE}\\trips_with_dynamic_co2.csv", index=False)
co2_by_vehicle.to_csv(f"{BASE}\\co2_by_vehicle.csv", index=False)
co2_by_driver.to_csv(f"{BASE}\\co2_by_driver.csv", index=False)
co2_by_route.to_csv(f"{BASE}\\co2_by_route.csv", index=False)

# ---------- PRINT SUMMARY ----------
print(f"Total CO2 emitted (kg) so far: {total_co2:.2f}")
print("\nTop 5 vehicles by CO2 emission:")
print(co2_by_vehicle.head())
print("\nTop 5 drivers by CO2 emission:")
print(co2_by_driver.head())
print("\nTop 5 routes by CO2 emission:")
print(co2_by_route.head())
```

```python
import pandas as pd
import numpy as np
from math import radians, sin, cos, asin, sqrt

class CO2Agent:
    def __init__(self, trips, vehicles, environment, fuel_preds=None):
        self.trips = trips.copy()
        self.vehicles = vehicles.copy()
        self.environment = environment.copy()
        self.fuel_preds = fuel_preds.copy() if fuel_preds is not None else None

        # CO2 factors per fuel type
        self.co2_factors = {'Petrol': 2.31, 'Diesel': 2.68, 'CNG': 2.74}

        # Preprocess data
        self._preprocess_data()

        # Storage for live segments
        self.live_segments = pd.DataFrame()

    # ----------------- INTERNAL -----------------
    @staticmethod
    def _haversine(lat1, lon1, lat2, lon2):
        """Vectorized Haversine distance in km"""
        lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
        dlon = lon2 - lon1
        dlat = lat2 - lat1
        a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
        return 2 * 6371 * np.arcsin(np.sqrt(a))

    def _preprocess_data(self):
        # Standardize columns
        for df in [self.trips, self.vehicles, self.environment]:
            df.columns = df.columns.str.strip().str.lower()
        if self.fuel_preds is not None:
            self.fuel_preds.columns = self.fuel_preds.columns.str.strip().str.lower()

        # Vehicles
        self.vehicles['baseline_kmpl'] = self.vehicles['baseline_kmpl'].fillna(10)
        self.vehicles['fuel_efficiency_l_per_km'] = 1 / self.vehicles['baseline_kmpl']
        self.vehicles['fuel_type'] = self.vehicles['fuel_type'].fillna('Petrol')

        # Trips
        self.trips['distance_km'] = self.trips['distance_km'].fillna(0)
        self.trips['vehicle_id'] = self.trips['vehicle_id'].fillna('Unknown')
        self.trips['driver_id'] = self.trips['driver_id'].fillna('Unknown')
        self.trips['route_id'] = self.trips['route_id'].fillna('Unknown')
        self.trips['anomaly_flag'] = self.trips.get('anomaly_flag', False)

        # Environment
        self.environment['fuel_modifier'] = self.environment['fuel_modifier'].fillna(1.0)
        if 'distance_km' not in self.environment.columns or self.environment['distance_km'].isna().any():
            self.environment['distance_km'] = self._haversine(
                self.environment['segment_start_lat'], self.environment['segment_start_lon'],
                self.environment['segment_end_lat'], self.environment['segment_end_lon']
            )

        # Merge vehicle info
        self.trips = self.trips.merge(
            self.vehicles[['vehicle_id','fuel_type','fuel_efficiency_l_per_km']],
            on='vehicle_id', how='left'
        )
        self.trips['fuel_type'] = self.trips['fuel_type'].fillna('Petrol')
        self.trips['fuel_efficiency_l_per_km'] = self.trips['fuel_efficiency_l_per_km'].fillna(0.1)

        # Merge fuel predictions
        if self.fuel_preds is not None:
            merge_cols = [c for c in ['assignment_id','driver_id','vehicle_id','route_id'] if c in self.fuel_preds.columns]
            self.trips = self.trips.merge(
                self.fuel_preds[merge_cols + ['predicted_fuel_l']],
                on=merge_cols, how='left'
            )
        else:
            self.trips['predicted_fuel_l'] = np.nan

    # ----------------- LIVE SEGMENTS -----------------
    def update_segment(self, trip_id, segment_update):
        """Add/update a live segment mid-trip."""
        segment_update['trip_id'] = trip_id
        if 'fuel_modifier' not in segment_update:
            segment_update['fuel_modifier'] = 1.0
        self.live_segments = pd.concat([self.live_segments, pd.DataFrame([segment_update])], ignore_index=True)
        return self.calculate_live_co2(trip_id)

    def calculate_live_co2(self, trip_id):
        """CO2 calculation for live trip segments."""
        trip = self.trips[self.trips['trip_id']==trip_id]
        segments = self.live_segments[self.live_segments['trip_id']==trip_id].copy()
        if segments.empty:
            return {'message':'No segments recorded yet'}

        trip_info = trip[['vehicle_id','driver_id','fuel_type','fuel_efficiency_l_per_km','predicted_fuel_l','anomaly_flag']].iloc[0]
        for col in trip_info.index:
            segments[col] = trip_info[col]

        # Fuel used
        segments['fuel_used'] = np.where(
            segments['predicted_fuel_l'].notna(),
            segments['predicted_fuel_l'],
            segments['distance_km'] * segments['fuel_efficiency_l_per_km'] * segments['fuel_modifier']
        )

        # CO2
        segments['co2_kg'] = np.where(
            segments['anomaly_flag'],
            0,
            segments['fuel_used'] * segments['fuel_type'].map(self.co2_factors).fillna(2.5)
        )

        total_co2 = segments['co2_kg'].sum()
        co2_by_vehicle = segments.groupby('vehicle_id')['co2_kg'].sum().to_dict()
        co2_by_driver = segments.groupby('driver_id')['co2_kg'].sum().to_dict()
        co2_by_route = segments.groupby('route_id')['co2_kg'].sum().to_dict()

        return {
            'trip_id': trip_id,
            'total_co2_kg': total_co2,
            'segments': segments[['route_id','distance_km','fuel_modifier','fuel_used','co2_kg']].to_dict(orient='records'),
            'co2_by_vehicle': co2_by_vehicle,
            'co2_by_driver': co2_by_driver,
            'co2_by_route': co2_by_route
        }

    # ----------------- FULL FLEET -----------------
    def calculate_total_co2(self):
        if self.live_segments.empty:
            return {'message':'No live segments recorded'}

        segments = self.live_segments.merge(
            self.trips[['trip_id','vehicle_id','driver_id','fuel_type','fuel_efficiency_l_per_km','predicted_fuel_l','anomaly_flag']],
            on='trip_id', how='left'
        )

        segments['fuel_used'] = np.where(
            segments['predicted_fuel_l'].notna(),
            segments['predicted_fuel_l'],
            segments['distance_km'] * segments['fuel_efficiency_l_per_km'] * segments['fuel_modifier']
        )
        segments['co2_kg'] = np.where(
            segments['anomaly_flag'],
            0,
            segments['fuel_used'] * segments['fuel_type'].map(self.co2_factors).fillna(2.5)
        )

        result = {}
        for trip_id, df in segments.groupby('trip_id'):
            result[trip_id] = {
                'total_co2_kg': df['co2_kg'].sum(),
                'segments': df[['route_id','distance_km','fuel_modifier','fuel_used','co2_kg']].to_dict(orient='records'),
                'co2_by_vehicle': df.groupby('vehicle_id')['co2_kg'].sum().to_dict(),
                'co2_by_driver': df.groupby('driver_id')['co2_kg'].sum().to_dict(),
                'co2_by_route': df.groupby('route_id')['co2_kg'].sum().to_dict()
            }
        return result

# ===============================
# DEMO
# ===============================
if __name__ == "__main__":
    trips = pd.DataFrame([
        {'assignment_id':'A1','trip_id':'T1','vehicle_id':'V1','driver_id':'D1','route_id':'R1','distance_km':100,'anomaly_flag':False},
        {'assignment_id':'A2','trip_id':'T2','vehicle_id':'V2','driver_id':'D2','route_id':'R2','distance_km':150,'anomaly_flag':False}
    ])
    vehicles = pd.DataFrame([
        {'vehicle_id':'V1','fuel_type':'Diesel','baseline_kmpl':12},
        {'vehicle_id':'V2','fuel_type':'Petrol','baseline_kmpl':10}
    ])
    environment = pd.DataFrame([
        {'route_id':'R1','segment_start_lat':12.0,'segment_start_lon':77.0,'segment_end_lat':13.0,'segment_end_lon':78.0,'fuel_modifier':1.1},
        {'route_id':'R2','segment_start_lat':14.0,'segment_start_lon':75.0,'segment_end_lat':15.0,'segment_end_lon':76.0,'fuel_modifier':0.9}
    ])
    fuel_preds = pd.DataFrame([
        {'assignment_id':'A1','driver_id':'D1','vehicle_id':'V1','route_id':'R1','predicted_fuel_l':8},
        {'assignment_id':'A2','driver_id':'D2','vehicle_id':'V2','route_id':'R2','predicted_fuel_l':15}
    ])

    agent = CO2Agent(trips, vehicles, environment, fuel_preds=fuel_preds)

    # Add live segments
    agent.update_segment('T1', {'route_id':'R1','distance_km':50})
    agent.update_segment('T1', {'route_id':'R1','distance_km':50})
    agent.update_segment('T2', {'route_id':'R2','distance_km':150})

    # Print CO2 outputs
    print("Trip T1 CO2:")
    print(agent.calculate_live_co2('T1'))
    print("\nTotal CO2 for all trips:")
    print(agent.calculate_total_co2())
```
